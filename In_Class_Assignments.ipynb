```python
COGS_RATE = 0.40  # 40% of sales

# Inputs
sales = float(input("Enter sales ($): "))
tax_rate_percent = float(input("Enter tax rate (%): "))

# Calculations
cogs = sales * COGS_RATE
gross_profit = sales - cogs
taxes = gross_profit * (tax_rate_percent / 100)
net_income = gross_profit - taxes

# Display (formatted like an income statement)
print("\nSales:           $ {:,.2f}".format(sales))
print("Cost's of Goods Sold: $ {:,.2f}".format(cogs))
print("Gross Profit:    $ {:,.2f}".format(gross_profit))
print("Taxes:           $ {:,.2f}".format(taxes))
print("Net Income:      $ {:,.2f}".format(net_income))
```

```python
import pandas as pd
import numpy as np

# --- Parameters ---
ticket_price = 10.0
initial_attendees = 20
fixed_costs = 200.0

# --- Advertising Amounts to Test ---
# Using numpy.arange to create a range of advertising dollars
# from 0 to 1000, with a step of 50
advertising_amounts = np.arange(0, 1001, 50)

# --- Lists to store results ---
results = []

# --- Calculate profit for each advertising amount ---
for ad_dollars in advertising_amounts:
    # Calculate additional attendees (round down to whole number as people can't be fractional)
    additional_attendees = int(2 * (ad_dollars**0.5))

    # Calculate total attendees
    total_attendees = initial_attendees + additional_attendees

    # Calculate revenue
    revenue = total_attendees * ticket_price

    # Calculate total costs
    total_costs = fixed_costs + ad_dollars

    # Calculate profit
    profit = revenue - total_costs

    # Store the results
    results.append({
        'Advertising Dollars': ad_dollars,
        'Additional Attendees': additional_attendees,
        'Total Attendees': total_attendees,
        'Revenue': revenue,
        'Total Costs': total_costs,
        'Profit': profit
    })

# --- Create DataFrame and display ---
df_profit = pd.DataFrame(results)

df_profit['Profit'] = df_profit['Profit'].map('${:,.2f}'.format)
df_profit['Revenue'] = df_profit['Revenue'].map('${:,.2f}'.format)
df_profit['Total Costs'] = df_profit['Total Costs'].map('${:,.2f}'.format)
df_profit['Advertising Dollars'] = df_profit['Advertising Dollars'].map('${:,.2f}'.format)

display(df_profit)
```
